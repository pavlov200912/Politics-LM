```python
import requests
from bs4 import BeautifulSoup

# Example of HTML with transcript
URL = 'http://www.kremlin.ru/events/president/transcripts/63854'
page = requests.get(URL)
page.raise_for_status()
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
paragraphs = []
for s in soup.find('div', {'class':'read__internal_content'}).find_all('p'):
    parent = s.parent
    if 'masha-ignore' in parent.find_parent('div').attrs['class']:
        continue
    paragraphs.append(s.text)
```

```python
text = ' '.join(paragraphs)
print(text)
```

```python
import re
def extract_persons(text):
    persons_and_texts = []
    copy_text = text
    exp = re.compile('[А-ЯЁ].[А-ЯЁ][а-яё]+:')
    current_person = ''
    while True:
        person_match = exp.search(copy_text)
        if not person_match:
            persons_and_texts.append((current_person, copy_text))
            break
        span = person_match.span()
        persons_and_texts.append((current_person, copy_text[:span[0]]))
        current_person = copy_text[span[0]:span[1]]
        copy_text = copy_text[span[1]:]
    if persons_and_texts[0] == ('', ''):
        persons_and_texts = persons_and_texts[1:]
    return persons_and_texts

extract_persons('В.Павлов: Стоит укреплять сельское хозяйство В.Путин: Согласен')
```

```python
for (p, t) in extract_persons(text):
    print(f'Person:{p}')
    print(t)
    print('___' * 15)
```
