```python
import re
def get_transcript_numbers(soup):
    links = []
    for link in soup.find_all('a'):
        links.append(link.get('href'))
    transcripts_numbers = set()

    #print(*links, sep='\n')
    for link in links:
        if link and re.match("/events/president/transcripts/[0-9]+$", link):
            transcripts_numbers.add(link.split('/')[-1])
    return transcripts_numbers
```

```python

import random
import time
def get_all_numbers(base_url='http://www.kremlin.ru/events/president/transcripts/page/'):
    NUMBER_OF_PAGES = 549
    numbers = []
    for page_number in range(1, NUMBER_OF_PAGES + 1):
        url = base_url + str(page_number)
        page = requests.get(url)
        page.raise_for_status()
        time.sleep(3)
        print(f'Got{page_number}')
        soup = BeautifulSoup(page.content, 'html.parser')
        numbers += get_transcript_numbers(soup)

    return numbers
```

```python
lst = get_all_numbers()
lst[:20]
```

```python

def makeRequestWithExponentialBackoff(url):
    for n in range(0, 5):
        answer = requests.get(url)
        if answer.status_code != 200:
            time.sleep((2 ** n) + random.random())
        else:
            print('Answered!')
            return  answer
    print('We\'re doomed')
```
